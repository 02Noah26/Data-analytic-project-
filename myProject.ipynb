## Data cleaning

```python
# we load our data that we will work on 
df = pd.read_csv('C:/Users/ismail/desktop/layoffs_data.csv')
df.head()
```

```python
# check number of rows and columns of the dataset 
df.shape
```

```python
df.info()
```

```python
#statistic values of the data 
df.describe()
```

```python
# check the number of unique values for each column 
df.nunique()
```

```python
# drop unuseful columns
df.drop(['List_of_Employees_Laid_Off', 'Date_Added','Source' ], axis=1, inplace= True)
```

```python
# calculate sum of null values for each column 
df.isnull().sum()
```

```python
# drop null values for each column 
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# sum the duplicated rows 
df.duplicated().sum()
```

```python
df[df.duplicated(keep =False)]
```

```python
#drop the duplicates 
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```
